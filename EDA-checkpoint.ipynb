```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
# Load dataset
def load_dataset(file_path): return pd.read_csv(file_path)
df = load_dataset("/Users/Chhanda/Documents/Preshna/EDA/insurance.csv") 
df.head(5)
```

```python
# Categorical data # countplot 
sns.countplot(data=df, x='sex') 
df['children'].value_counts()
```

```python
# pie chart 
df['sex'].value_counts().plot(kind='pie',autopct='%0.2f')
```

```python
# Numberical Value # Histogram treat as categorical data distribution of data 
plt.hist(df['age'])
```

```python
# Distplot # probability density function --> The line is kde # skweness 
sns.distplot(df['age'])
```

```python
# Boxplot # 5 number summary # median , percentile, min=Q1-1.5*IQR, max=Q3+1.5*IQR, IQR = (Q1-Q3), outlier 
sns.boxplot(df['age'])
```

```python
df.describe()
```

```python
# Scatter plot(Numerical - Numerical) for the relation between two column 
sns.scatterplot(x='bmi', y='charges',hue='sex', style='smoker',size='children' ,data=df)
plt.show()
```

```python
# Bar plot (Numerical - Categorical) # average age in children 
sns.barplot(x='children',y='age',hue='sex',data=df)
```

```python
#Box plot (Numerical - Categorical) 
sns.boxplot(x='sex',y='age',hue='children',data=df)
```

```python
# Displot 
sns.distplot(df[df['children']==0]['age'],hist=False) 
sns.distplot(df[df['children']==1]['age'],hist=False)
```

```python
# HeatMap (Categorical - Categorical) 
sns.heatmap(pd.crosstab(df['age'],df['children']))
```

```python
# pairplot 
sns.pairplot(data=df,hue='children')
```

```python
sns.lineplot(x='age',y='children',data=df)
```
